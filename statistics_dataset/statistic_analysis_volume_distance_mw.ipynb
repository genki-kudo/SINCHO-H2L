```python
#calculate max-distance in 3Dmolecule
import numpy as np
with open('selected_comps_of_ligandexpo.sdf','r')as inp:
    dist_list = []
    label_count = 0
    atom_count = 5000
    num = 0
    comp_xyz = {}
    name_list = []
    name_count = 0
    for line in inp:
        if '>  <Field_1>' in line:
            name_count = 1
            continue
        if name_count == 1:
            name_list.append(line[0:3])
            name_count = 0
            continue
        if 'Schrodinger Suite 2020-3.' in line:
            label_count = 1
            continue
        if label_count == 1:
            atom_count= int(line.split()[0])
            num = 0
            label_count = 0
            comp_xyz = {}
            continue
        if atom_count != 5000:
            ls = line.split()
            comp_xyz[num] = np.array([float(ls[0]), float(ls[1]), float(ls[2])])
            num += 1
            atom_count-= 1
        if atom_count == 0:
            atom_count = 5000
            num = 0
            dist = 0
            for i in comp_xyz:
                for j in comp_xyz:
                    if i == j:
                        continue
                    elif i>j:
                        dist_tmp = float(np.linalg.norm(comp_xyz[i] - comp_xyz[j]))
                        if dist_tmp >= dist:
                            dist = dist_tmp
                            atom1 = i
                            atom2 = j
            dist_list.append([dist, atom1, atom2])
            
        
```

```python
print(len(dist_list))
print(len(name_list))
lst = []
for i in range(len(dist_list)):
    lst.append([name_list[i], dist_list[i][0], dist_list[i][1], dist_list[i][2]])
print(lst)
```

```python
def est_mw_ind(comp,spath, sbl_list):
    #spath = Chem.GetShortestPath(mol, 5, 10)
    
    atom = mol.GetAtoms()
    print(atom)
    est_mw_ind = 0
    for a in atom:
        if a.GetIdx() in spath[1:-1]:
            #print(a.GetMass())
            sbl_list.append(a.GetSymbol())
            est_mw_ind+=a.GetMass()
    return(est_mw_ind)
```

```python
#calculate dist-passmw 
from rdkit import Chem
from matplotlib import pyplot as plt

comps = Chem.SDMolSupplier('selected_comps_of_ligandexpo.sdf')
dist_l = []
estmw_l = []
d_mw_l = []
num = 0
avail = 0
sbl_list = []
for mol in comps:
    if mol is None:
        num += 1
    else:
        spath = Chem.GetShortestPath(mol, dist_list[num][1], dist_list[num][2])
        pass_mw = est_mw_ind(mol,spath,sbl_list)
        #print("distance-> ",dist_list[num][0])
        #print("est_mw-> ", pass_mw, )
        dist_l.append(dist_list[num][0])
        estmw_l.append(pass_mw)
        d_mw_l.append([name_list[num], dist_list[num][0], pass_mw])
        num += 1
        avail += 1
res1 = np.polyfit(np.array(dist_l), np.array(estmw_l),1)
x = np.linspace(5,10,40)
y = np.poly1d(res1)(x)

fig = plt.figure(figsize=(12,8))
plt.scatter(dist_l, estmw_l)
plt.xlim(0,40)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Max-Distance [Å]",fontsize=25)
plt.ylabel("MW of shortest-path",fontsize=25)
plt.show()
fig.savefig("distance_passmw.png")

fig = plt.figure(figsize=(12,8))
plt.scatter(dist_l, estmw_l)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Max-Distance [Å]",fontsize=25)
plt.ylabel("MW of shortest-path",fontsize=25)
plt.show()
fig.savefig("distance_passmw_nolimit.png")

print(res1[0], res1[1])
fit = []
fit.append(res1[0])
fit.append(res1[1])

import json

open('distance_passmw.json','w').write(json.dumps(d_mw_l))
open('d_mw_fitting.json','w').write(json.dumps(fit))
d1 = json.loads(open('distance_passmw.json').read())
d2 = json.loads(open('d_mw_fitting.json').read())
print(len(d1))
d2
```

```python
#calculate volume-mw 
from rdkit import Chem
from matplotlib import pyplot as plt
from rdkit.Chem import Descriptors
from rdkit.Chem import AllChem

comps = Chem.SDMolSupplier('selected_comps_of_ligandexpo.sdf')
vol_l = []
mw_l = []
vol_mw_l = []
num = 0
avail = 0
sbl_list = []
for mol in comps:
    #print(mol)
    if mol is None:
        num += 1
    else:
        mw = Descriptors.HeavyAtomMolWt(mol)
        vol = AllChem.ComputeMolVolume(mol, gridSpacing=1.0)
        #print("volume-> ",vol)
        #print("est_mw-> ", pass_mw)
        vol_l.append(vol)
        mw_l.append(mw)
        vol_mw_l.append([name_list[num], vol, mw])
        num += 1
        avail += 1

res1 = np.polyfit(np.array(vol_l), np.array(mw_l),1)
x = np.linspace(5,10,40)
y = np.poly1d(res1)(x)
fig = plt.figure(figsize=(12,8))
plt.scatter(vol_l, mw_l)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Volume [Å^3]",fontsize=25)
plt.ylabel("MW",fontsize=25)
plt.show()
fig.savefig("vol_mw.png")
print(res1[0], res1[1])
fit = []
fit.append(res1[0])
fit.append(res1[1])

import json

open('volume_mw.json','w').write(json.dumps(vol_mw_l))
open('vol_mw_fitting.json','w').write(json.dumps(fit))
d1 = json.loads(open('volume_mw.json').read())
d2 = json.loads(open('vol_mw_fitting.json').read())
print(len(d1))
d2
```
