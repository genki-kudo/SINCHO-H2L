```python
from rdkit import Chem
from matplotlib import pyplot as plt
from rdkit.Chem import Descriptors
from rdkit.Chem import AllChem
from rdkit.Chem import rdMolTransforms

comps = Chem.SDMolSupplier('selected_comps_of_ligandexpo.sdf')
alert_symb = ['Se', 'Rh', 'Zn', 'B', 'Pb', 'Ti', 'Ca', 'Al', 'Sn', 'Re', 'Ag',
              'Sc', 'Be', 'V', 'Pt', 'Ga', 'Br', 'Ni', 'Fe', 'F', 'Li', 'Cr', 
              'Hg', 'Sb', 'Mg', 'I', 'Na', 'Si', 'Mn', 'Cl', 'Pr', 'Pd', 'Ru', 
              'As', 'Cu', 'W', 'U', 'Te', 'Au', 'Mo', 'Ir']
dist_list = []
passmw_list = []
vol_list = []
mw_list = []
num = 0
avail = 0
for mol in comps:
    #error-avoid
    if mol is None:
        continue
    #object preparation
    atom = mol.GetAtoms()
    bonds = mol.GetBonds()
    conf = mol.GetConformer()
    atomnum = mol.GetNumAtoms()
    
    #CONPS check
    ch1 = 0
    for a in atom:
        if a.GetSymbol() in alert_symb:
            ch1 += 1
            break
    if ch1 ==1:
        num += 1
        continue
    #single-molecule check
    if '.' in Chem.MolToSmiles(mol):
        num += 1
        continue
    #bond-length check
    ch2 = 0
    for bond in bonds:
        bl = rdMolTransforms.GetBondLength(conf, bond.GetBeginAtomIdx(), bond.GetEndAtomIdx())
        if bl >=2.5:
            ch2 += 1
            break
    if ch2 ==1:
        num += 1
        continue
    #property append
    dist = 0
    num += 1
    avail += 1
    for a in range(atomnum):
        for b in range(atomnum):
            if rdMolTransforms.GetBondLength(conf,a,b) >= dist:
                atom1 = a
                atom2 = b
                dist = rdMolTransforms.GetBondLength(conf,a,b)
    dist_list.append(dist)
    spath = Chem.GetShortestPath(mol, atom1, atom2)
    pass_mw = 0
    for a in atom:
        if a.GetIdx() in spath[1:-1]:
            pass_mw += a.GetMass()
    passmw_list.append(pass_mw)
    vol_list.append(AllChem.ComputeMolVolume(mol, gridSpacing=0.2))
    mw_list.append(Descriptors.HeavyAtomMolWt(mol))
    
print(dist_list)
print(passmw_list)
print(vol_list)
print(mw_list)

print(num)
print(avail)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import json
import pandas as pd

res1 = np.polyfit(np.array(vol_list), np.array(mw_list),1)
x = np.linspace(5,10,40)
y = np.poly1d(res1)(x)
fig = plt.figure(figsize=(12,8))
plt.scatter(vol_list, mw_list)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Volume [Å^3]",fontsize=25)
plt.ylabel("MW",fontsize=25)
plt.show()
fig.savefig("vol_mw.png")
print(res1[0], res1[1])
fit = []
fit.append(res1[0])
fit.append(res1[1])
s1 = pd.Series(vol_list)
s2 = pd.Series(mw_list)
cor = s1.corr(s2)
print(cor)

open('vol_mw_fitting.json','w').write(json.dumps(fit))

res1 = np.polyfit(np.array(dist_list), np.array(passmw_list),1)
x = np.linspace(5,10,40)
y = np.poly1d(res1)(x)
fig = plt.figure(figsize=(12,8))
plt.scatter(dist_list, passmw_list)
plt.xlim(0,40)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Max-Distance [Å]",fontsize=25)
plt.ylabel("MW of shortest-path",fontsize=25)
plt.show()
fig.savefig("distance_passmw.png")
print(res1[0], res1[1])
fit = []
fit.append(res1[0])
fit.append(res1[1])
s1 = pd.Series(dist_list)
s2 = pd.Series(passmw_list)
cor = s1.corr(s2)
print(cor)

open('dist_passmw_fitting.json','w').write(json.dumps(fit))

res1 = np.polyfit(np.array(dist_list), np.array(mw_list),1)
x = np.linspace(5,10,40)
y = np.poly1d(res1)(x)
fig = plt.figure(figsize=(12,8))
plt.scatter(dist_list, mw_list)
plt.xlim(0,40)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Max-Distance [Å]",fontsize=25)
plt.ylabel("MW",fontsize=25)
plt.show()
fig.savefig("distance_mw.png")
print(res1[0], res1[1])
fit = []
fit.append(res1[0])
fit.append(res1[1])
s1 = pd.Series(dist_list)
s2 = pd.Series(mw_list)
cor = s1.corr(s2)
print(cor)

open('mw.json','w').write(json.dumps(mw_list))
open('maxdist.json','w').write(json.dumps(dist_list))
open('passmw.json','w').write(json.dumps(passmw_list))
open('volume.json','w').write(json.dumps(vol_list))

```
